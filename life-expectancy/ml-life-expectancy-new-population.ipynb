```python
import pandas as pd

df = pd.read_csv("LifeExpectancyData_NewPopulation.csv")
df
```

```python
df.drop(["Country", "Year"], axis=1, inplace=True)
df.dropna(inplace=True)
df = df[df['percentage_expenditure']!= 0]
df
```

```python
df.loc[df['Status'] == "Developing", "Status"] = 0
df.loc[df['Status'] == "Developed", "Status"] = 1
df
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.2)

X = train.drop(["Life_expectancy"], axis=1)
Y = train["Life_expectancy"]
X
```

```python
from sklearn.ensemble import RandomForestRegressor
clf = RandomForestRegressor()

clf.fit(X, Y)
```

```python
test_x = test.drop(["Life_expectancy"], axis=1)
test_y = test["Life_expectancy"]
predict_y = clf.predict(test_x)
predict_y
```

```python
import math
mse = test_y-predict_y
print(mse)
mse = (mse * mse)
print("mse = ",mse.mean())
print("rmse = ", math.sqrt(mse.mean()))
```

```python
from matplotlib import pyplot as plt

mse.plot.hist(grid=True, bins=20, rwidth=0.9,
                   color='#607c8e')
plt.title('MSE histogram')
plt.xlabel('mean square error')
plt.ylabel('Counts')
plt.grid(axis='y', alpha=0.75)
```

```python
import seaborn as sns
from matplotlib import pyplot as plt

features = list(X.columns)
feature_imp = pd.Series(clf.feature_importances_,index=features).sort_values(ascending=True)
#print(feature_imp)

# Creating a bar plot
sns.barplot(x=feature_imp, y=feature_imp.index)

# Add labels to graph
plt.xlabel('Feature Importance Score by Random Forest')
plt.ylabel('Features')
plt.legend()
plt.show()
```
