```python
import pandas as pd

df = pd.read_csv("LifeExpectancyData.csv")
df
```

```python
df.drop(["Country", "Year"], axis=1, inplace=True)
df.dropna(inplace=True)
df
```

```python
df.loc[df['Status'] == "Developing", "Status"] = 0
df.loc[df['Status'] == "Developed", "Status"] = 1
df
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.2)

X = train.drop(["Life expectancy "], axis=1)
Y = train["Life expectancy "]
X
```

```python
from sklearn.ensemble import RandomForestRegressor
clf = RandomForestRegressor()

clf.fit(X, Y)
```

```python
test_x = test.drop(["Life expectancy "], axis=1)
test_y = test["Life expectancy "]
predict_y = clf.predict(test_x)
predict_y
```

```python
mse = test_y-predict_y
mse = (mse * mse).mean()
mse
```

```python
import seaborn as sns
from matplotlib import pyplot as plt

features = list(X.columns)
feature_imp = pd.Series(clf.feature_importances_,index=features).sort_values(ascending=True)
#print(feature_imp)

# Creating a bar plot
sns.barplot(x=feature_imp, y=feature_imp.index)

# Add labels to graph
plt.xlabel('Feature Importance Score by Random Forest')
plt.ylabel('Features')
plt.legend()
plt.show()
```
